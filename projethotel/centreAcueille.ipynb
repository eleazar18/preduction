```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import joblib
```

```python
#chargement du dataset
dataset=pd.read_csv("future.csv")
dataset
```

```python
#determiner x et y
x=dataset[['niv_class','servent','pt_dejeunne','souppee','nbr_jrs']]
y=dataset['prix']
```

```python
#on subdivise notre dataset en 2 parties training data et test data en tenant compte des x et y
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
#choisir l'algorithme
regresseur= LinearRegression()
```

```python
#entrainer grace a la methode fit
regresseur.fit(x_train,y_train)

#Enregistrement du model
joblib.dump(regresseur,'notreModel.pkl')
```
